```python
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def load_df(file_name):
    df = pd.read_csv(file_name).set_index('log regularization parameter')
    return df[df['Distribution'] == 'Gaussian']
```

# linear regression

```python
# isotropic
n, d = 300, 350
df_isotropic = load_df('Isotropic_mis_specified_n=' + str(n) + '_d=' + str(d) + '_Ridge.csv')

# junk and harmful features
n, d = 300, 3000
df_junk = load_df('Junk_feature_mis_specified_n=' + str(n) + '_d=' + str(d) + '_Ridge.csv')
df_harmful = load_df('Non_benign_mis_specified_n=' + str(n) + '_d=' + str(d) + '_Ridge.csv')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(20, 5))

df_isotropic['Risk bound'] *= 1.03
sns.lineplot(data = df_isotropic, ci = 95, ax = axes[0])
axes[0].set_title('Isotropic: n = 300, d = 350', fontsize = 14)
axes[0].set_xlim((15, -2.5))
axes[0].set_xlabel('log ridge parameter', fontsize = 12)
axes[0].legend(loc = 'upper left', fontsize = 11)

df_junk['Risk bound'] *= 1.03
sns.lineplot(data = df_junk, ci = 95, ax = axes[1])
axes[1].set_title('Junk feature: n = 300, d = 3000', fontsize = 14)
axes[1].set_ylim((-0.2, 5.5))
axes[1].set_xlim((12, -2.5))
axes[1].set_xlabel('log ridge parameter', fontsize = 12)
axes[1].legend(loc = 'upper left', fontsize = 11)

df_harmful['Risk bound'] *= 1.03
sns.lineplot(data = df_harmful, ci = 95, ax = axes[2])
axes[2].set_title('Harmful feature: n = 300, d = 3000', fontsize = 14)
axes[2].set_ylim((-0.2, 5.5))
axes[2].set_xlim((10, -10))
axes[2].set_xlabel('log ridge parameter', fontsize = 12)
axes[2].legend(loc = 'upper left', fontsize = 11)

plt.show()
```

# Linear classification

```python
n, d = 100, 120
df_isotropic = load_df('Isotropic_n=' + str(n) + '_d=' + str(d) + '_l2_margin.csv')

n, d = 100, 2000
df_junk = load_df(file_name = 'Junk_feature_n=' + str(n) + '_d=' + str(d) + '_l2_margin.csv')
df_harmful = load_df('Non_benign_n=' + str(n) + '_d=' + str(d) + '_l2_margin.csv')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(20, 5))

df_isotropic['Risk bound'] *= 1.03
sns.lineplot(data = df_isotropic, ci = 95, ax = axes[0])
axes[0].set_title('Isotropic: n = 100, d = 120', fontsize = 14)
axes[0].set_ylim((-0.05, 1.5))
axes[0].set_xlim((15, -2.5))
axes[0].set_xlabel('log regularization', fontsize = 12)
axes[0].legend(loc = 'upper left', fontsize = 11)

df_junk['Risk bound'] *= 1.01
sns.lineplot(data = df_junk, ci = 95, ax = axes[1])
axes[1].set_title('Junk feature: n = 100, d = 2000', fontsize = 14)
axes[1].set_ylim((-0.05, 1.5))
axes[1].set_xlim((12.5, -5))
axes[1].set_xlabel('log regularization', fontsize = 12)
axes[1].legend(loc = 'upper left', fontsize = 11)

df_harmful['Risk bound'] *= 1.02
sns.lineplot(data = df_harmful, ci = 95, ax = axes[2])
axes[2].set_title('Harmful feature: n = 100, d = 2000', fontsize = 14)
axes[2].set_ylim((-0.05, 1.5))
axes[2].set_xlim((10, -10))
axes[2].set_xlabel('log regularization', fontsize = 12)
axes[2].legend(loc = 'upper left', fontsize = 11)

plt.show()
```
